```kotlin
import kotlin.math.pow
import kotlin.random.Random
%use dataframe
%use kandy
```

```kotlin
fun runGame(
  moneyA: Int,
  moneyB: Int,
  winProbA: Double,
): Pair<Map<Int,Int>, Int> {
  var moneyA = moneyA
  var moneyB = moneyB
  val playerAWinHistory = mutableMapOf<Int, Int>()
  var turn = 0
  var totalWins = 0

  while (!(moneyA == 0 || moneyB == 0)) {
    Random.nextDouble().let {
      if (it < winProbA) {
        moneyA -= 1
        totalWins++
      } else {
        moneyB -= 1
      }
      playerAWinHistory[turn] = totalWins
    }
    turn++
  }
  return playerAWinHistory to turn
}
```

```kotlin
val winProbabilitiesPlayerA = listOf(0.2,0.5,0.8)
val games = winProbabilitiesPlayerA.zip(1..3).map { (winProbA, gameNumber) ->
  val (history, turns) = runGame(moneyA = 50, moneyB = 50, winProbA = winProbA)
  print(history)
  history.map { (turn, wins) -> Triple(turn, wins, gameNumber) }
}.flatten()
```

```kotlin
val df = dataFrameOf(
  "Turn" to games.map { it.first },
  "Wins" to games.map { it.second },
  "Game" to games.map { "Game ${it.third}" }
)
df.plot {
  line {
    x("Turn")
    y("Wins")
    color("Game")
  }
  layout {
    title = "Cumulative Wins per Game"
  }
}
```
