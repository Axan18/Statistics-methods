```kotlin
import kotlin.math.pow
import kotlin.random.Random
%use dataframe
%use kandy
```

```kotlin
fun runGame(
    moneyA: Int,
    moneyB: Int,
    winProbA: Double,
): Int {
    var moneyA = moneyA
    var moneyB = moneyB
    var turn = 0
    while (!(moneyA == 0 || moneyB == 0)) {
        Random.nextDouble().let {
            if (it < winProbA) {
                moneyB -= 1
                moneyA += 1
            } else {
                moneyA -= 1
                moneyB += 1
            }
        }
        turn++
    }
    return turn
}
```

```kotlin
val moneyA = 50
val moneyB = 50
val winProbabilitiesA = (1..100).map { it / 100.0 }

val longestGames = winProbabilitiesA.map { winProbA ->
  List(1000) { runGame(moneyA, moneyB, winProbA) }.max()
}
```

```kotlin
val df = dataFrameOf(
  "Win probability of player A" to winProbabilitiesA,
  "Longest games" to longestGames
)

df.plot {
  line {
    x("Win probability of player A")
    y("Longest games")
  }
  layout.title = "Longest game for given win probability of player  A"
  layout.size = 700 to 400
}
```
